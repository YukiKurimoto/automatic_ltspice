# 並列回路

```python
import pandas as pd
import matplotlib.pyplot as plt
from standingwave_class import Standingwave
```

```python
parallel = Standingwave('test_case_files/test_parallel.json')
```

```python
# parallel.run_all_tests()
```

```python
parallel.get_dfs()
parallel.t_cut(0.0049)
parallel.cut_standingwave(0.01, 1000)
parallel.plot_dfs()
```

```python
amp_dict = parallel.get_amp_means()
amp_dict
```

```python
amp_df = pd.DataFrame({'test_name': amp_dict.keys(), 'amp': amp_dict.values()})
```

```python
amp_df['turns'] = 0
for i in range(10):
    amp_df.iat[i, 2] = i+1
for i in range(10,19):
    amp_df.iat[i, 2] = (i-8)*10
```

```python
amp_df.plot(x='turns', y = 'amp')
```

```python
# 巻き数1~10
amp_df.iloc[0:10,:].plot(y='amp')
```

```python
# 巻き数10~100
amp_df.iloc[10:20,:].plot(y='amp')
```
