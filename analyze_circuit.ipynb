# 並列回路

```python
import pandas as pd
import matplotlib.pyplot as plt
from standingwave_class import Standingwave
from test_case_files import test_parallel as parallel_config
from test_case_files import test_series as series_config
```

```python
parallel = Standingwave(parallel_config.test_file, parallel_config.test_case)
```

```python
# parallel.run_all_tests()
```

```python
parallel.get_dfs()
parallel.t_cut(0.0049)
parallel.cut_standingwave(0.01, 1000)
parallel.plot_dfs()
```

```python
amp_dict = parallel.get_amp_means()
```

```python
induct_list = []
amp_list = []
for key in amp_dict.keys():
    induct_list.append(float(key.replace("L=", "").replace("u", "")))
    amp_list.append(amp_dict[key])
induct_amp_df = pd.DataFrame({'inductance': induct_list, 'amp': amp_list})
```

```python
plt.figure()
induct_amp_df.plot(x = "inductance", y = 'amp')
```

```python
plt.savefig('images/parallel_inductance_amp.png')
plt.close('all')
```
