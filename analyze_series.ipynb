# 巻き数それぞれの出力電圧の振幅のリストを取得する

```python
change_list = {}
N_amp_dict = {}
for i in range(10):
    series = Standingwave('test_case_files/test_series.json')
    change_list['L'] = change_L[i]
    change_list['R_L'] = change_R_L[i]
    for r in change_R:
        change_list['R'] = r
        series.change_element_df_save(change_list)
    series.t_cut(0.0049)
    series.cut_standingwave(0.01, 1000)
    N_amp_dict[f"{str(i+1)}"] = series.get_amp_means()
    del series
    gc.collect()
for i in range(10,19):
    series = Standingwave('test_case_files/test_series.json')
    change_list['L'] = change_L[i]
    change_list['R'] = change_R[i]
    for r in change_R:
        change_list['R'] = r
        series.change_element_df_save(change_list)
    series.t_cut(0.0049)
    series.cut_standingwave(0.01, 1000)
    N_amp_dict[f"{str((i-8)*10)}"] = series.get_amp_means()
    del series
    gc.collect()
N_amp_dict
```

```python
R_amp_dict = {}
for turns in N_amp_dict.keys():
    temp = []
    for i in range(len(change_R)):
        test = list(N_amp_dict[turns]['V(test)'].keys())[i]
        temp.append(N_amp_dict[turns]['V(test)'][test])
    R_amp_dict[turns] = temp
```

```python
for turns in R_amp_dict.keys():
    R_amp_df = pd.DataFrame({"R":change_R, "amp":R_amp_dict[turns]})
    R_amp_df.plot(x="R", y="amp")
```

### 巻き数1~10

```python
for turns in R_amp_dict.keys():
    R_amp_df = pd.DataFrame({"R":change_R, "amp":R_amp_dict[turns]})
    R_amp_df.iloc[0:10,:].plot(x="R", y="amp", title=f"Turns per layer = {turns}")
```

### 巻き数10~100

```python
# 巻き数10~100
for turns in R_amp_dict.keys():
    R_amp_df = pd.DataFrame({"R":change_R, "amp":R_amp_dict[turns]})
    R_amp_df.iloc[10:20,:].plot(x="R", y="amp", title=f"Turns per layer = {turns}")
```

### 出力電圧が最高だったテストパターン

```python
max_turns = ''
max_test = ''
max_V = 0.0
for turns in N_amp_dict.keys():
    for test in N_amp_dict[turns]['V(test)'].keys():
        v = N_amp_dict[turns]['V(test)'][test]
        if v>max_V:
            max_turns = turns
            max_V = v
            max_test = test
print(f"{max_turns}回巻: {max_test}")
```

# 巻き数ごとにRとシミュレーション期間全体の消費電力の平均の関係のリスト

```python
change_list = {}
N_W_dict = {}
all_case_len = len(change_L)*len(change_R_L)*len(change_R)*20
for i in range(10):
    series = Standingwave('test_case_files/test_series.json')
    change_list['L'] = change_L[i]
    change_list['R_L'] = change_R_L[i]
    W_temp = {}
    for r in change_R:
        change_list['R'] = r
        series.change_element_df_save(change_list, get_value="I(R2)")
        W_mean = series.get_W(R_name='R1', R_value=float(change_list['R']), change_list=change_list)['W(R1)'].mean()
        W_temp[get_test_name(change_list)] = W_mean
    N_W_dict[f"{str(i+1)}"] = W_temp
    del series
    gc.collect()
    percent = (100*len(change_L)*len(change_R_L)*len(change_R)*(i+1))//all_case_len
    sys.stdout.write("\r%dパーセント完了" % percent)
    sys.stdout.flush()
for i in range(10,19):
    series = Standingwave('test_case_files/test_series.json')
    change_list['L'] = change_L[i]
    change_list['R'] = change_R[i]
    W_temp = {}
    for r in change_R:
        change_list['R'] = r
        series.change_element_df_save(change_list, get_value="I(R2)")
        W_mean = series.get_W(R_name='R1', R_value=float(change_list['R']), change_list=change_list)['W(R1)'].mean()
        W_temp[get_test_name(change_list)] = W_mean
    series.t_cut(0.0049)
    series.cut_standingwave(0.01, 1000)
    N_W_dict[f"{str((i-8)*10)}"] = W_temp
    del series
    gc.collect()
    percent = (100*len(change_L)*len(change_R_L)*len(change_R)*(i+1))//all_case_len
    sys.stdout.write("\r%dパーセント完了" % percent)
    sys.stdout.flush()
N_W_dict
```
